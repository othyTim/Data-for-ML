#Business Insights

```python
ax = sns.countplot(data=df_insure2, hue='OUTCOME', x='VEHICLE_OWNERSHIP');
ax.set_title('Number of Vehicle ownership and Outcome status');
```

Vehicle Ownership:
- 0 = not owned
- 1 = Owned

Outcome:
- 0 = Not claimed Loan
- 1 = Claimed Loan

The Graph above shows us that there is a strong negeative relation between Ownership stauts and our Target outcome: it shows us that increase in Owned vehicles would have a increase in Not claimed loans, which mean decrease in claimed loans.

```python
ax = sns.histplot(binwidth=1,
            hue=df_insure2['OUTCOME'],
            x=df_insure2['INCOME'],
            data=df_insure2,
            stat="count",
            multiple="dodge");
ax.set_title('Outcome Distribution amongst Income Classes');
```

The graph above shows us that there is a trend in regard to the income classes and there loan claim status.

The higher the class the less claimed loans. it shows us that the only class that has more claimed loans than non claimed is the poverty class. it also shows us the higher the class the greater the proportional gap of the clamed and non claimed loans.
