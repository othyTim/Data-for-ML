##Adult Income Dataset

###Load describe data

```python
#Load data
dataset ='/content/drive/MyDrive/CodingDojo/02-MachineLearning/Week08/Data/adult.csv'
df_adult_income = pd.read_csv(dataset)

#Describe data
df_adult_income.info()
df_adult_income.head()
```

```python
#copy of df for my own sanity
df_income = df_adult_income.copy()
```

###Dropping columns

```python
drop_income = df_income[{'race'}]
```

```python
df_income = df_income.drop(columns = drop_income)
df_income
```

```python
rename_dict = {'native-country' : 'native_country'}
rename_dict
```

```python
df_income = df_income.rename(rename_dict,axis=1)
```

###Check for Duplicates

```python
#Checking Duplication count
df_income.duplicated().sum()
```

```python
#Identify all duplicated rows (including the first)
duplicated_rows_all = df_income.duplicated(keep=False)
duplicated_rows_all.sum()
```

```python
df_income[duplicated_rows_all].sort_values('fnlwgt')
```

```python
# Drop duplicates (but keep the first!)
df_income = df_income.drop_duplicates()
# Obtain the sum of duplicated to confirm all have been dropped
df_income.duplicated().sum()
```

```python
null_values2 = df_income.isna().sum()
null_values2
```

###Identified and corrected inconsistencies

```python
data_types2 = df_income.dtypes
object_data_types2 = data_types2[(data_types2 == "object")]
object_data_types2
```

```python
object_data_types2.index
```

```python
## Loop through the index of object_data_types
for column in object_data_types2.index:
    # Display the unique values from the column
    print(column)
    print(df_income[column].unique())
    print('\n')
```

Check counts of "?" values

```python
df_income['workclass'].value_counts()
```

```python
df_income['occupation'].value_counts()
```

```python
df_income['native_country'].value_counts()
```

Based on the number, I will drop the rows with "?"

```python
#dropping rows with "?" value
df_income = df_income[df_income.workclass != '?']
df_income = df_income[df_income.occupation != '?']
df_income = df_income[df_income.native_country != '?']
```

```python
#In comparison with the original data removing the "?" we have lost less than 10% of data which is okay based on that our dataset is so big
df_income.info()
```

###Visuals

```python
ax = sns.histplot(binwidth=1,
            hue=df_income['income'],
            x=df_income['gender'],
            data=df_income,
            stat="count",
            multiple="dodge");
ax.set_title('Income Distribution amongst Males and Females');
```

```python
# Seaborn countplot
ax = sns.countplot(data=df_income, x='relationship')
```

Graph above indicates relationship status of "Husband" is the majority of this data set
